# Step 3: continue with the rest. 

```python
pd.set_option('display.max_rows', None)
  
MAX_TO_SHOW = 5
 
def get_candidates(user_input, maximum_to_show = MAX_TO_SHOW):
    user_input = user_input.strip() 
    candidates = DF_COLLEGE[DF_COLLEGE["College"].str.contains(user_input, case=False, flags=0, na=None, regex=True)]
    candidates = candidates[["Rank", "College", "Town", "Wiki", "Web"]].copy() # , "Summary"
    candidates["No."]  = [i + 1 for i in range(candidates.shape[0])] 
    candidates = candidates[["No.", "Rank", "College", "Town", "Wiki", "Web"]]
    return candidates.head(maximum_to_show)
```

# overall flow

```python
first_msg = "Please enter your college name. Enter 'Q' to exit"
repeat_msg = "No College Found. Please Search Again.Enter 'Q' to exit"
 
candidates = pd.DataFrame() 
is_first_search = True

while candidates.empty:
    if is_first_search:  
        user_input = input(first_msg)
        is_first_search = False
    else:
        user_input = input(repeat_msg) 
     
    candidates = get_candidates(user_input, maximum_to_show = MAX_TO_SHOW)
    
    if not candidates.empty:
        break 

display(HTML(candidates.to_html(index=False))) 

college_no_min, college_no_max = 1, candidates.shape[0]


college_no = -1 

while college_no == -1:
    college_no = input("Please Enter College No.")
    if college_no.isnumeric():
        college_no = int(college_no)
        if college_no < college_no_min:
            college_no = -1 
        if college_no > college_no_max:
            college_no = -1
    else: 
        college_no = 0
        break 
        
if college_no > 0:    
    college_no = college_no - 1 ## Index 
    college_row = candidates.iloc[college_no] 

    college = college_row["College"]
    town = college_row["Town"] 

    display_msg = f"<font size = 5, color = red > You Have Chosen {college} at {town}. Generating Summeries.... </font>"
    display(Markdown(display_msg))

    wiki = college_row["Wiki"]
    wiki = wiki.split("/")[-1]
    wiki_summary = wikipedia.summary(wiki) 
    display(wiki_summary)
```
